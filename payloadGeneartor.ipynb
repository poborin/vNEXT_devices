```
open System.Text.Json

type Da = {
    id: string
    Name: string
    location: string
    ``type``: string
}

let idPrefix = "DVID"
let locations = ["Melbourne"; "Geelong"; "Ballarat"; "Bendigo"; "Melton"; "Shepparton"; "Wangaratt"; "Yea"]
let types = ["Preassure Sensor"; "Temperature Sensor"; "CO2 Sensor"; "Humidity Sensor"]

let shuffleR (r : Random) xs = xs |> Seq.sortBy (fun _ -> r.Next())

let devices = Seq.init 1000 (fun x -> 
                    let n = sprintf "%08i" x
                    let id = sprintf "DVID%s" n
                    {
                        id = id;
                        Name = sprintf "device %i name" x;
                        location = locations |> shuffleR (System.Random ()) |> Seq.head
                        ``type`` = types |> shuffleR (System.Random ()) |> Seq.head
                    }
                )

let payload = {| correlationid = "27280ecb-4db5-4a5d-b542-175d324ce173";
                devices = devices
                |}

let json = JsonSerializer.Serialize payload
json
```
